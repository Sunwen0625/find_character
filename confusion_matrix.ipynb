```python
from fastai.vision.all import *
learn = load_learner(r"D:\program\py\illegal_stop\test\predict\model.pkl")

dls = ImageDataLoaders.from_folder(
        r'D:\program\py\illegal_stop\test\predict\augment_positive_images',
        valid_pct=0.2,
        seed=42,
        item_tfms=Resize(224),
        batch_tfms=aug_transforms(), 
        num_workers=0, # 增強也可以開
    )
learn.dls = dls
# 準確度
learn.validate()

# 混淆矩陣
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()

# 隨機預測幾張
learn.show_results()
```
